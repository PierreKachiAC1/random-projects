```python
import pandas as pd
import matplotlib.pyplot as plt

dataset=pd.read_csv("jobs_in_data.csv")
dataset.head
```

```python
mean = dataset.groupby('job_category')['salary'].mean()
plt.figure(figsize=(10, 6))
mean.plot(kind='bar', color='skyblue')
plt.xlabel('Job Category')
plt.ylabel('Mean Salary')
plt.title('Mean Salary by Job Category')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(dataset['experience_level'], dataset['salary'], color='skyblue', alpha=0.5)
plt.xlabel('Experience Level')
plt.ylabel('Salary')
plt.title('Salary vs. Experience Level')
plt.grid(True)
plt.show()
```

```python
mean_salary_by_employment_type = dataset.groupby('employment_type')['salary'].mean()

plt.figure(figsize=(10, 6))
mean_salary_by_employment_type.plot(kind='bar', color='skyblue')
plt.xlabel('Employment Type')
plt.ylabel('Mean Salary')
plt.title('Mean Salary by Employment Type')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
#i did use chatgpt but fehmen kel shi

experience_level_mapping = {'Entry-level': 1, 'Mid-level': 2, 'Senior': 3, 'Executive': 4}#turning experience levels into numerical values la ne2dir nesta3meloun la ntalli3 corelation

dataset['experience_level_numeric'] = dataset['experience_level'].map(experience_level_mapping)

correlation_coefficient = dataset['experience_level_numeric'].corr(dataset['salary'])#generating the corelation

#the results are bad but i am blaming the dataset saraha
plt.figure(figsize=(10, 6))
plt.scatter(dataset['experience_level_numeric'], dataset['salary'], color='skyblue', alpha=0.5)

plt.xlabel('Experience Level (Numeric)')
plt.ylabel('Salary')
plt.title('Salary vs. Experience Level (Correlation: {:.2f})'.format(correlation_coefficient))
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#ok ma ha kazzib im starting to believe ha2 3laye w mesh 3al dataset

slope, intercept = np.polyfit(dataset['experience_level_numeric'], dataset['salary'], 1)

# Plot the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(dataset['experience_level_numeric'], dataset['salary'], color='skyblue', alpha=0.5, label='Data')

plt.plot(dataset['experience_level_numeric'], slope * dataset['experience_level_numeric'] + intercept, color='red', label='Linear Regression')

plt.xlabel('Experience Level (Numeric)')
plt.ylabel('Salary')
plt.title('Salary vs. Experience Level with Linear Regression')
plt.legend()
plt.grid(True)
plt.show()
```
